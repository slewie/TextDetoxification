# Final solution

For final solution, I use T5-base-detox model. I will finetune it on 15 epochs on bigger part of the dataset that for the experiment. I uploaded this model to [huggingface](https://huggingface.co/slewie/t5-ultradetox)

```python
import sys
sys.path.append("TextDetoxification")
from src.utils.trainer import Trainer
from src.data.make_dataloader import _make_dataloader_transformers
```

```python
SEED = 0
SIM_MODEL_PATH = 'TextDetoxification/models/fasttext.bin'
SAMPLE_SIZE = 20000
DATA_PATH = 'data/interim/tox_sentences.csv'
NUM_EPOCHS = 15
MODEL_PATH = './models/'
```

## Model creation

```python
model_name = "slewie/t5-ultradetox"

trainer = Trainer(model_name, 'transformers', random_seed = SEED, sim_model_path = SIM_MODEL_PATH)
tokenized_dataset = _make_dataloader_transformers(DATA_PATH, model_name, sample_size=SAMPLE_SIZE)
```

```python
trainer.train(num_epochs=NUM_EPOCHS, tokenized_dataset=tokenized_dataset, save_model=True, batch_size=8, model_path=MODEL_PATH)
```
